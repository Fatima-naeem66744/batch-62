# Problem : High Low

```python
import random

num_rounds = 5
print("Welcome to the High Low Game")
print("---------------------------------------------------------------")

while True:
    score = 0
    round_num = 1
    while round_num <= num_rounds:
        print(f"Round {round_num}")
        user_num = random.randint(1, 100)
        computer_num = random.randint(1, 100)
        print(f"Your number is {user_num}")

        prompt = input("Guess if it is higher or lower than the computer's number (type 'higher' or 'lower'):\t")
        
        if prompt == "higher":
            if user_num > computer_num:
                score += 1
                print(f"You were right! The computer's number was {computer_num}. Your score: {score}")
            else:
                score = max(0, score - 1) 
                print(f"You were wrong! The computer's number was {computer_num}. Your score: {score}")
            round_num += 1  
            print("---------------------------------------------------------------")

        elif prompt == "lower":
            if user_num < computer_num:
                score += 1
                print(f"You were right! The computer's number was {computer_num}. Your score: {score}")
            else:
                score = max(0, score - 1)  
                print(f"You were wrong! The computer's number was {computer_num}. Your score: {score}")
            round_num += 1  
            print("---------------------------------------------------------------")

        else:
            print("Invalid input! The game will restart from Round 1.")
            score = 0
            round_num = 1
            print("---------------------------------------------------------------")
            continue  

        print("\n-------------------------")
        
    if round_num > num_rounds:
        print(f"\nGame Over! Your final score is {score}")
        play_again = input("Do you want to play again? (yes/no): ")
        if play_again.lower() != "yes":
            break
```

```python
import random

class HighLowGame:
    def __init__(self, num_rounds=5):
        self._num_rounds = num_rounds   # Private attribute
        self._score = 0                 # Private attribute
        self._round_num = 1             # Private attribute

    def _update_score_and_round(self, correct_guess):
        """Private method to update score and round number."""
        if correct_guess:
            self._score += 1
            print(f"You were right! Your score: {self._score}")
        else:
            self._score = max(0, self._score - 1)
            print(f"You were wrong! Your score: {self._score}")
        self._round_num += 1

    def play_round(self):
        print(f"Round {self._round_num}")
        user_num = random.randint(1, 100)
        computer_num = random.randint(1, 100)
        print(f"Your number is {user_num}")

        prompt = input("Guess if it is higher or lower than the computer's number (type 'higher' or 'lower'):\t")

        if prompt == "higher":
            correct_guess = user_num > computer_num
            print(f"The computer's number was {computer_num}.")
            self._update_score_and_round(correct_guess)

        elif prompt == "lower":
            correct_guess = user_num < computer_num
            print(f"The computer's number was {computer_num}.")
            self._update_score_and_round(correct_guess)

        else:
            print("Invalid input! Restarting from Round 1.")
            self._score = 0
            self._round_num = 1

        print("---------------------------------------------------------------")

    def play_game(self):
        print("Welcome to the High Low Game")
        print("---------------------------------------------------------------")

        while True:
            self._score = 0
            self._round_num = 1

            while self._round_num <= self._num_rounds:
                self.play_round()

            print(f"\nGame Over! Your final score is {self._score}")
            play_again = input("Do you want to play again? (yes/no): ")
            if play_again.lower() != "yes":
                print("Thank you for playing! Goodbye!")
                break

# Instantiate and start the game
game = HighLowGame()
game.play_game()
```
