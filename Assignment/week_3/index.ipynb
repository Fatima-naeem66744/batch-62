```python
guests = ["Alice", "Bob", "Charlie", "Diana"]

# Printing initial invitations
for guest in guests:
    print(f"Dear {guest}, you are invited to dinner.")

# Message about only being able to invite two people
print("\nUnfortunately, I can only invite two people for dinner.")

# Removing guests until only two remain
while len(guests) > 2:
    removed_guest = guests.pop()
    print(f"Sorry {removed_guest}, I can't invite you to dinner.")

# Messages to the remaining guests
for guest in guests:
    print(f"Dear {guest}, you are still invited to dinner.")

# Remove the last two names from the list
del guests[0]
del guests[0]

# Printing the final state of the guest list
print("\nFinal guest list:", guests)
```

```python
places = ['Tokyo', 'Barcelona', 'New York', 'Sydney', 'Cape Town']
print(places)
#sorted()
print(sorted(places))
#original
print(places)
#reverse alphabetic order
print(sorted(places,reverse=True))
#original
print(places)
#reverse
places.reverse()
print(places)
#reverse again to get the original order
places.reverse()
print(places)
#sort
places.sort()
print(places)
places.sort(reverse=True)
print(places)
```

```python
languages = ["English", "Spanish", "Mandarin", "French", "German"]
print(languages)
languages.append("korean")
print(languages)
languages.insert(1,"Turkish")
print(languages)
languages.sort()
print(languages)
languages.sort(reverse=True)
print(languages)
languages.reverse()
print(languages)
languages.pop()
print(languages)
languages.remove("English")
print(languages) 
print(languages.count("German"))
print(languages.index("German"))
languages.clear()
print(languages)
```
