```python
character = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
character[0:4]
character[-4:-2]
character[:4]
character[:]
character[-4:-2:-1]   # -1 step mean Python will return an empty list because slicing starts at the index -4 and tries to move backwards but stops immediately at -2.


```

```python
character = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

character[::-1]  ## in opposite direction
```

```python
character = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
character[-2:-5:-1]
```

```python
num = [4,7,88,6]
num1 = [3,4,5,66]
num.append(num1)
print(num)
```

```python
num = [4,7,88,6]
print(num.index(4))
```
