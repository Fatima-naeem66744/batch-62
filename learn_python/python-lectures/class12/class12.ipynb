## create your own error

```python
class student_card():
    def __init__(self, name, age, grade):
        if age < 18 or age > 60:
            raise studentError("You are not eligible")
        self.name = name
        self.age = age
        self.grade = grade

class studentError(Exception):    
    pass

student = student_card("Fatima", 15, "A+")
print(student.name, student.age, student.grade)
```

```python
class student_card():
    def __init__(self, name, age, grade):
        if age < 18 or age > 60:
            raise studentError("You are not eligible")
        self.name = name
        self.age = age
        self.grade = grade

class studentError(Exception):    
    pass
try:
  student = student_card("Fatima", 15, "A+")  
  print(student.name, student.age, student.grade)
except Exception as e:
    print(f"StudentError : {e}")  

```

```python
with open("./abc1.txt" ,  "r") as file:
   print(file.write("hello"))
```

```python
with open("./abc1.txt" ,  "w") as file:
   print(file.read())
   print(file.write("hello"))
```

```python
with open("./abc1.txt" ,  "w") as file:# it will remove your previous text and write the new text.
   print(file.write("hello"))
```

```python
with open("./abc1.txt","r+") as file: # it has the power of reading and writing both but it reads 
# the  previous file and enter the new words but do not read the new words 
   print(file.read())
   print(file.write("hello world")) 
```

```python
with open("./abc1.txt","r+") as file: 
   print(file.read())
   print(file.write("hello world")) 
   print(file.seek(39))
   print("After : " , file.read())
```
