### shallow copy
 Modifications to nested objects affect both the original and the copy because they share references to the same nested objects.
### deep copy
 Modifications to nested objects only affect the copy, not the original, because all nested objects are independently copied.

# IMP NOTE:
1. we can use the module `copy` if we dont want to use it then we should use it with list slicing and the method `list`


```python
original = [1, 2, 3, [4, 5]]
shallow_copy = original[:]  # Creates a shallow copy

shallow_copy[0] = 10        # Modifying shallow_copy's first element
shallow_copy[3][0] = 40     # Modifying a nested element in the shallow copy

print("Original:", original)         # Output: [1, 2, 3, [40, 5]]
print("Shallow Copy:", shallow_copy) # Output: [10, 2, 3, [40, 5]]
```

```python
original = ["apple", "banana", ["cherry", "date"]]
shallow_copy = list(original)  # Creates a shallow copy

shallow_copy[1] = "blueberry"   # Changes only shallow_copy
shallow_copy[2][0] = "cranberry"  # Modifies a nested element

print("Original:", original)         # Output: ["apple", "banana", ["cranberry", "date"]]
print("Shallow Copy:", shallow_copy) # Output: ["apple", "blueberry", ["cranberry", "date"]]
```

```python
import copy
original = [1, 2, [3, 4], [5, [6, 7]]]
shallow_copy = list(original) 
# Modify the shallow copy
shallow_copy[2][0] = 99
shallow_copy[3][1][0] = 100
print(original) 
print(shallow_copy)
```

```python
# Define a recursive function to create a deep copy
def deep_copy(data):
    if isinstance(data, list):
        return [deep_copy(item) for item in data]  # Recursively copy each item if it's a list
    else:
        return data  # Return the item directly if it's not a list (like integers or strings)

# Original list with nested structure
original = [1, 2, [3, 4], [5, [6, 7]]]

# Use the recursive function to create a deep copy
deep_copied = deep_copy(original)

# Modify the deep copy and check if it affects the original
deep_copied[2][0] = 30  # Change an element in a nested list
deep_copied[3][1][0] = 60  # Change a deeply nested element

print("Original:", original)        # Output: [1, 2, [3, 4], [5, [6, 7]]]
print("Deep Copied:", deep_copied)   # Output: [1, 2, [30, 4], [5, [60, 7]]]
```

## deep copy


```python
# Define a simple recursive deep copy function (for general cases)
def deep_copy(data):
    if isinstance(data, list):
        return [deep_copy(item) for item in data]  # Recursively copy each item in the list
    else:
        return data  # Return the item directly if it's not a list

# Original list
names = ["fatima", "tehreem" , ["areeba" , "jimin"]]

# Create a deep copy of the list
name = deep_copy(names)

# Modify the deep copy
name[0] = "ali"
name[2][0] = "jhope"

# Outputs
print("Original:", names)     # Output: ["fatima", "tehreem"]
print("Deep Copy:", name)     # Output: ["ali", "tehreem"]

# Modify the deep copy and show it's independent of the original
```

### difference b/w shallow copy or deep copy

```python
names = ["fatima", "tehreem", ["nested", "list"]]  # A list with a nested list
name = names[:]  # Shallow copy
print("Original:", names)
print("Shallow Copy:", name)

name[0] = "ali"  # Change a top-level element in the shallow copy
name[2][0] = "changed"  # Change an element in the nested list (shared between original and shallow copy)

print("Modified Shallow Copy:", name)   # Output: ["ali", "tehreem", ["changed", "list"]]
print("Original after modification:", names)  # Output: ["fatima", "tehreem", ["changed", "list"]]
```

```python
# Define a deep copy function
def deep_copy(data):
    if isinstance(data, list):
        return [deep_copy(item) for item in data]  # Recursively copy each item
    else:
        return data  # Return the item directly if it's not a list

names = ["fatima", "tehreem", ["nested", "list"]]  # A list with a nested list
name = deep_copy(names)  # Deep copy
print("Original:", names)
print("Deep Copy:", name)

name[0] = "ali"  # Change a top-level element in the deep copy
name[2][0] = "changed"  # Change an element in the nested list

print("Modified Deep Copy:", name)  # Output: ["ali", "tehreem", ["changed", "list"]]
print("Original after modification:", names)  # Output: ["fatima", "tehreem", ["nested", "list"]]
```

###  Shallow Copy:
 When you modify the nested list inside the shallow copy (name[2][0] = "changed"), it also modifies the nested list in the original (names) because the shallow copy doesn't create a new copy of the nested list. It just references the original nested list.

### Deep Copy:
 When you modify the nested list inside the deep copy (name[2][0] = "changed"), the original names list remains unchanged. This is because deep_copy recursively creates copies of all nested lists, so the nested list inside name is completely separate from the one inside names.
