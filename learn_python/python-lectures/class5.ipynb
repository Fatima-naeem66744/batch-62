```python
database: List[Tuple[str, str]] = [
    ("sir qasim", "234"),
    ("sir zia", "568"),
    ("sir wasim", "989")
]

for row in database:
    username , password = row
    print(username , password )
```

```python
database: List[Tuple[str, str]] = [
    ("sir qasim", "234"),
    ("sir zia", "568"),
    ("sir wasim", "989")
]
user = input("write the user name:")
input_pass = input("write the password:")

for row in database:
    username , password = row
    if user == username and input_pass == password:
        print("valid user")
else:
        print("invalid user")
    
```
