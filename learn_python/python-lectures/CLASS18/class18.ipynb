### concatenate 

```python
import pandas as pd
df = pd.concat([ss1 , ss2]).reset_index(drop = True)
display(df.info())       # gives you the whole information about your object
print("==============================")
display(df.describe())
print("==============================")
display(df.head())
print("==============================")
display(df.head())
print("==============================")
display(df.tail())
```

```python
import pandas as pd
import numpy as np
total = 1000
s1 = np.random.randint(80 , 100 , (total , 2))
ss1 : pd.DataFrame = pd.DataFrame(s1 , columns=['s1' , 's2'])

s2 = np.random.randint(60 , 80 , (total , 2))
ss2 : pd.DataFrame = pd.DataFrame(s2 , columns=['s1' , 's2'])

df = pd.concat([ss1 , ss2]).reset_index(drop = True)
df.to_csv("students.csv" , index=False)

df = pd.read_csv("students.csv")

print(df.head())
```

## Apply custom function on each column

```python
def grade(per):
    if per >= 90:
        return 'A'
    elif per >= 80:
        return 'B'
    elif per >= 70:
        return 'C'
    elif per >= 60:
        return 'D'
    else:
        return 'F'
grade(60)    
```

## how to apply the function on the columns 

```python
df['percentage'].head().apply(grade)
```
