methods of dictionaries

```python
from typing import Dict, Union

key = str
value = Union[list, set, Dict, int]
data: Dict[key, value] = {"fname": "M.Qasim"}

# Use get method correctly
print(data.get("pakistan" , "N.A"))  # This will print "NA" since "pakistan" is not a key in the dictionary.
print(data)
```

```python
from typing import Dict, Union

key = str
value = Union[list, set, Dict, int]
data: Dict[key, value] = {"fname": "M.Qasim" , 
                          "name" : "Sir zia"}


print(data.values())  #give the value
print(data.keys())   #give the keys
print(data.items())   #give the combination  of values and keys
for k,v in data.items():
    print(k , v)

```
