```python
import pandas as pd
from typing import Any


student : dict[str , list[Any]] = {
    "roll no " : [1,2,3],
    "name": ["John" , "ali" , "areeba"],
    "age": [20 ,21,34],
    "education": ["master" ,"master", "master"]
}

# Create a DataFrame
df = pd.DataFrame(student)
print(student)

# Print the DataFrame
print(df)
```

```python
alien = {
    "position": 0,
    "y-position": 25,
    "speed": "slow"
}

original = print(f"Original position: {alien['position']}")

# Compare the value of the "speed" key in the alien dictionary
if alien["speed"] == "slow":
    position = 1
elif alien["speed"] == "medium":
    position = 2
else:
    position = 3

alien["position"] += position
print(f"New position is: {alien['position']}")
```

```python
favorite_languages = {
    'jen': ['python', 'rust'],
    'sarah': ['c'],
    'edward': ['rust', 'go'],
    'phil': ['python', 'haskell'],
}

# Loop through the dictionary
for name, languages in favorite_languages.items():
    print(f"\n{name.title()}'s favorite languages are:")
    
    # Loop through the list of languages for each person
    for language in languages:
        print(f"\t{language.title()}")
```
