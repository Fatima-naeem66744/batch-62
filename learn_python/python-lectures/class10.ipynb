#### generator function
- iterate on element one by one
- stop after each iteration
- remember old iteration value


````python
# abc means abstract base classes
from collections.abc import Iterator
def my_range(start : int , end : int , step = 1) -> Iterator[int]:
     for i in range(start , end , step):
         yield i
# A generator function in Python is defined like a normal function,
 #  but whenever it needs to generate a value, 
 # it does so with the yield keyword rather than return.

a  : Iterator[int] = my_range(0,11)
print(next(a)) 
print(next(a))  
print(type(a))
print(a)

````

````python
for i in a:
    print(i)
````

## Decorator function
### syntax

````python


def decorator_name(func):
    def wrapper(*args, **kwargs):
        # Code to execute before calling the function
        result = func(*args, **kwargs)
        # Code to execute after calling the function
        return result
    return wrapper

@decorator_name
def my_function():
    # Function code
    pass


````

````python
from typing import Callable

def decorator(func: Callable):
    def my_func():
        print("this function is called")
        func()
        print("this function is called after the above function")
    return my_func

@decorator
def say_hello():
    print("hello")

say_hello()
````

````python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

# Calling the decorated function
say_hello()
````

````python
from typing import Callable

def decorator(func: Callable):
    def wrapper(num1):
        print("this function is called")
        func(num1)
        print("this function is called after the above function")
    return wrapper

@decorator
def say_hello(num1):
    print(num1)
say_hello(100)
````

## recursive functions
* Recursive functions are functions that calls itself.


````python
def factorial(x):
    if x == 1:
        return 1
    else:
        return x * factorial(x-1)
print(f"the factorial is {factorial(3)}")  
    
````

````python
def fibonacci(n):
    # Base cases
    if n == 0:
        return 0
    elif n == 1:
        return 1
    # Recursive case
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Example usage
print(fibonacci(6))  # Output: 8
````
