# Callable
* The main purpose of using the __call__ method in Python is to allow an instance of a class to be used as if it were a function. 

```python
class Adder:
    def __init__(self, x):
        self.x = x

    def __call__(self, y):
        return self.x + y

add_five = Adder(4)
print(add_five(3))  # Output: 8
# Here, add_five behaves like a function, even though it's an object of the Adder class.
print(add_five.__call__(7)) # otherwise
```

## Why Use super().__init__()?
- When you use inheritance, the child class can extend or modify the behavior of the parent class. By using `super()`
 - It ensures that any initialization that happens in the parent class (like setting up attributes) is not lost when the child class defines its own __init__ method.

```python
from abc import ABC , abstractmethod
from traitlets import Bool

# abstract base classes
class animal(ABC):
    @abstractmethod
    def __init__(self) -> None:
        #super().__init__()
        self.living_thing : bool = True 
       

fatima = animal() # we cannot make an object directly   
fatima.living_thing
```

## abstract method
* Abstraction means hiding the complicated stuff and only showing what is necessary. It makes using objects easier by not showing all the details of how they work.

* Think of it as a TV remote. You press buttons to change channels, but you don’t need to know the technology inside

```python
from abc import ABC, abstractmethod

class Shape(ABC):  # ABC stands for Abstract Base Class
    @abstractmethod
    def calculate_area(self):
        pass

# Shape is an abstract base class (ABC) that defines a method calculate_area() using the @abstractmethod
# decorator from the abc module.
# An abstract method is a method that is declared but contains no implementation.
# It must be overridden in any subclass that inherits from Shape.

import math

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def calculate_area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

# Usage example
circle = Circle(5)
rectangle = Rectangle(3, 4)

print("Area of circle:", circle.calculate_area())    
```

```python
from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def __init__(self) -> None:
        self.living_thing: bool = True
    
    @abstractmethod
    def eat(self, food):
        pass

# Concrete class (plants)
class Plants(Animal):
    def __init__(self, name):
        super().__init__()  # Initialize the base class
        self.name = name
    
    # Implement the abstract method 'eat'
    def eat(self, food):
        print(f"{self.name} absorbs {food} as nutrients.")
    
    # Define the __call__ method
    def __call__(self):
        print(f"{self.name} grows in the forest")

# Create an instance of the Plants class
plant = Plants("rose")
plant()  # Call the __call__ method
plant.eat("sunlight")  # Call the implemented eat method
```
