# Callable
* The main purpose of using the __call__ method in Python is to allow an instance of a class to be used as if it were a function. 

```python
class Adder:
    def __init__(self, x):
        self.x = x

    def __call__(self, y):
        return self.x + y

add_five = Adder(5)
print(add_five(3))  # Output: 8
# Here, add_five behaves like a function, even though it's an object of the Adder class.
```

```python
from typing import Any


class factory():
    def __init__(self , exponent = 2):
        self.exponent = exponent
    def __call__(self , base) -> Any:
        return base ** self.exponent
a = factory()
a.exponent
```

```python
a(5) # a acts like the function here 
```

```python
from typing import Any

class Factorial:
    def __init__(self):
        self.cache = {0: 1, 1: 1}

    def __call__(self, number: int) -> int:
        if number not in self.cache:
            self.cache[number] = number * self(number - 1)
        return self.cache[number]

result = Factorial()
print(result.cache) 
print(result(5))

print(result.cache) 
```
