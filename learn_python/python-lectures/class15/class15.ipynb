# Callable
* The main purpose of using the __call__ method in Python is to allow an instance of a class to be used as if it were a function. 

```python
class Adder:
    def __init__(self, x):
        self.x = x

    def __call__(self, y):
        return self.x + y

add_five = Adder(5)
print(add_five(3))  # Output: 8
# Here, add_five behaves like a function, even though it's an object of the Adder class.
```

```python
from typing import Any


class factory():
    def __init__(self , exponent = 2):
        self.exponent = exponent
    def __call__(self , base) -> Any:
        return base ** self.exponent
a = factory()
a.exponent
```

```python
from typing import Any

class Factorial:
    def __init__(self):
        self.cache = {0: 1, 1: 1}

    def __call__(self, number: int) -> int:
        if number not in self.cache:
            self.cache[number] = number * self(number - 1)
        return self.cache[number]

result = Factorial()
print(result.cache) 
print(result(5))

print(result.cache) 
```

## Why Use super().__init__()?
- When you use inheritance, the child class can extend or modify the behavior of the parent class. By using `super()`
 - It ensures that any initialization that happens in the parent class (like setting up attributes) is not lost when the child class defines its own __init__ method.

```python
from abc import ABC , abstractmethod
from traitlets import Bool

# abstract base classes
class animal(ABC):
    @abstractmethod
    def __init__(self) -> None:
        #super().__init__()
        self.living_thing : bool = True 
       

fatima = animal() # we cannot make an object directly   
fatima.living_thing
```

```python
from abc import ABC, abstractmethod

# Abstract base class
class animal(ABC):
    @abstractmethod
    def __init__(self) -> None:
        super().__init__()
        self.living_thing: bool = True


class cat(animal):
    def __init__(self) -> None:  
        super().__init__()       
        

fatima = cat() 

print(fatima.living_thing)  
```
