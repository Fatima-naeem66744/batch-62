## create your own error

```python
class student_card():
    def __init__(self, name, age, grade):
        if age < 18 or age > 60:
            raise studentError("You are not eligible")
        self.name = name
        self.age = age
        self.grade = grade

class studentError(Exception):    
    pass

student = student_card("Fatima", 15, "A+")
print(student.name, student.age, student.grade)
```

## file handling

```python
import os

# Check the existence of the file you intend to read
file_path = "abc.txt"

if os.path.exists(file_path):
    with open(file_path, "r") as file:
        print(file.read())
else:
    print("File does not exist!")
```
