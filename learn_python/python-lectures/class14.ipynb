##### over-riding concept

```python
class mother():
    def __init__(self, name):
        self.name = name
        self.eyecolor = "blue"

    def speaking(self , words ):
        print(f"Mother speaks {words}")
       

class father():
    def __init__(self, name):
        self.name = name
        self.height = "6 feet"

    def speaking(self , words ):
        print(f"Father speaks {words}")    

class child(father , mother):
    def __init__(self, father_name , mother_name , child_name):
        father.__init__(self , father_name)
        mother.__init__(self , mother_name)
        self.child_name = child_name

fatima = child("Naeem" , "saba" , "Fatima Naeem" )     
  
print(f"fatima height : {fatima.height}")
print(f"fatima eye color : {fatima.eyecolor}")
fatima.speaking("Pakistan Zindabad")
```

# function overloading
- One function perform different tasks 

```python
from typing import overload, Union

@overload
def add(x: int, y: int) -> int:
    ...

@overload
def add(x: float, y: float) -> float:
    ...

@overload
def add(x: str, y: str) -> str:
    ...

def add(x: Union[str, float, int], y: Union[str, float, int]) -> Union[str, float, int]:
    if isinstance(x, int) and isinstance(y, int):
        return x + y
    elif isinstance(x, float) and isinstance(y, float):
        return x + y
    elif isinstance(x, str) and isinstance(y, str):
        return x + y
    else:
        raise TypeError("Unsupported types for addition")
add(7, 4)
```

# Method overloading

```python
from typing import overload, Union

class Adder:
    @overload
    def add(self, x: int, y: int) -> int:
        ...

    @overload
    def add(self, x: float, y: float) -> float:
        ...

    @overload
    def add(self, x: str, y: str) -> str:
        ...

    def add(self, x: Union[str, float, int], y: Union[str, float, int]) -> Union[str, float, int]:
        if isinstance(x, int) and isinstance(y, int):
            return x + y
        elif isinstance(x, float) and isinstance(y, float):
            return x + y
        elif isinstance(x, str) and isinstance(y, str):
            return x + y
        else:
            raise TypeError("Unsupported types for addition")


adder_instance = Adder()

result = adder_instance.add(7, 4)
print(result) 
```

## overriding 

- In Python, `polymorphisms` refer to the occurrence of something in multiple forms. As part of polymorphism, a Python child class has methods with the same name as a parent class method. This is an essential part of programming.

```python
class animal():
    def eating(self , food):
        print(f"Animal is eating {food}")
class birds(animal):
    def eating(self , food):
        print(f"Bird is eating {food}")  

bird = birds()
print(bird.eating("bread"))     

animal = animal()
animal.eating("grass")
```

- Create a class Engine with a method start(), and a class Wheels with a method rotate(). Create a class Car that inherits from both Engine and Wheels. Instantiate Car and call both start() and rotate() methods.

```python
class Engine:
    def start(self):
        print("Engine started")

class Wheels:
    def rotate(self):
        print("Wheels rotating")

class Car(Engine, Wheels):
    def __init__(self):
        
        print("Car created")

    def result(self):
        self.start()   
        self.rotate()  

my_car = Car()
my_car.result()
```

```python
from typing import overload
class calculator():
    @overload
    def add(self, a: int, b: int) -> int:
        """Add two integers"""
        return a + b
    @overload
    def add(self,a: int, b: int, c : int) -> int:
        """Add three integers"""
        return a + b + c
    def add(self, *args):   
        return sum(args)
cal = calculator()
cal.add(2,4,8)    
   
```

- Create a class Shape with a method area() that returns 0. Create subclasses Circle and Rectangle that override the area() method to return the correct area for the respective shapes.



```python
import math
class shape():
    def area(self):
        return 0
class circle(shape):
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius ** 2
class Rectangle(shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Override the area method
    def area(self):
        return self.length * self.width    
Circle = circle(5)
print(f"Area of the circle: {Circle.area()}")
rectangle = Rectangle(4, 6)
print(f"Area of the rectangle: {rectangle.area()}")
```
