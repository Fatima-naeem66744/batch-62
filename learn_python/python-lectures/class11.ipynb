## pass by value

```python
a : int = 5  # immutable  data type
def abc(num_1 : int):
    print("The value of the number is" , num_1)
    num_1 = 3
    print("The value of number is" , num_1)
abc(a) # pass by value 
print(a)    
```

### pass by reference

```python
a : list = [1,2,3,4]  #mutable means changeable data type 
print(id(a))
def abc(num_1 : list[int]):
    print(id(num_1))
    num_1.append(100)
    print(f"The value of number is {num_1} and address is {id(num_1)}")
abc(a)
print(a)    
```

```python
a : list = [1,2,3,4]  #mutable data type 
def abc(num_1 : list[int]):
    num_1 = [8]
    num_1.append(100)
    print("The value of number is" , num_1)
abc(a)
print(a)    
```
