````python
class teacher():
    def __init__(self , teacher_name : str , teacher_id : int) -> None:
      self.name : str = teacher_name  # atrributes
      self.id : int = teacher_id
      self.oraganization : str = "PUNJAB GROUP OF COLLEGES"

    def speak(self , words) -> None:  
       print(f"{self.name} says '{words}'")

    def teach(self , subject) -> None:    
       print(f"{self.name} is teaching {subject}")

obj1 = teacher("Sir Mansoor" , 1)
obj2 = teacher("Sir Qasim" , 2) # We can say that the object is the instance of class teacher()

print(obj1.name)
print(obj1.oraganization)
print(obj1.id)
print(obj1.teach("Computer"))
print(obj1.speak("SDLC is the process every software engineer must know"))


print(obj2.name)
print(obj2.oraganization)
print(obj2.id)
print(obj2.teach("GENERATIVE AI"))




````

# inheritance
```
class class_name1
class class_name(class_name1):
   pass

```

````python
class parents():
    def __init__(self):
        self.color = "fair"
        self.haircolor = "brown"
        
    def speak(self , words):
        print(f"speak method {words}")

    def watch(self , object):
        print(f"You look like {object}")    

class  child(parents):
     pass

## parent class
obj = parents()
print(obj.color)   
print(obj.haircolor)
obj.speak("hello")
obj.watch("a boy")

## child class
obj1 = child()
print(obj1.color)   
print(obj.haircolor)
obj.speak("hello guys")
obj.watch("a girl")
````

 ## another example of inheritance

````python
class parents():
    def __init__(self):
        self.color = "fair"
        self.haircolor = "brown"
        
    def speak(self , words):
        print(f"speak method {words}")

    def watch(self , object):
        print(f"You look like {object}")    

class  child(parents):
     def teaching(self , subject):
         print(f"Method od teaching : {subject} ")

## parent class
obj = parents()
print(obj.color)   
print(obj.haircolor)
obj.speak("hello")
obj.watch("a boy")

## child class
obj1 = child()
print(obj1.color)   
print(obj1.haircolor)
obj.speak("hello guys")
obj.watch("a girl")
obj1.teaching("Islamiyat")
````

## real life example

````python
class employee:
    def __init__(self, name: str) -> None:
        self.name = name
        self.education = ''
        self.department = 'department no. 1'

class designer(employee): 
    def __init__(self, title: str, name: str) -> None:
        super().__init__(name) 
        self.title = title

class developer(employee): 
    def __init__(self, title: str, name: str) -> None:
        super().__init__(name) 
        self.title = title
        self.programming_skills: list[str] = ["python"]  


designer1 = designer("Designing and Cloth Designing", "Alice")
developer1 = developer("Programming and Making Applications", "Bob")


print(f"{designer1.name} specializes in {designer1.title}")
print(f"{developer1.name} has programming skills in {developer1.programming_skills}")
print(designer1.department)
````

````python
class Courses:
    def __init__(self, course_name, duration, instructor, credits, level) -> None:
        self.course_name = course_name
        self.duration = duration
        self.instructor = instructor
        self.credits = credits
        self.level = level

    def details(self):
        print(f"Course: {self.course_name}, Duration: {self.duration} months, Instructor: {self.instructor}, Credits: {self.credits}, Level: {self.level}")


course1 = Courses("Mathematics", 6, "Miss Sadia", 3, "Intermediate")
course2 = Courses("Physics", 5, "Mr. Ahmed", 4, "Advanced")
course3 = Courses("English Literature", 4, "Miss Ayesha", 2, "Beginner")
course4 = Courses("Chemistry", 6, "Mr. Usman", 3, "Intermediate")
course5 = Courses("Biology", 5, "Miss Maria", 3, "Advanced")


course1.details()
course2.details()
````

````python
class Classroom:
    def __init__(self, room_number, capacity, subject, teacher) -> None:
        self.room_number = room_number
        self.capacity = capacity
        self.subject = subject
        self.teacher = teacher

    def class_info(self):
        print(f"Classroom {self.room_number} is for {self.subject} and is taught by {self.teacher}. Capacity: {self.capacity} students.")

# Creating classroom instances
classroom1 = Classroom("101", 30, "Maths", "Miss Sadia")
classroom2 = Classroom("202", 25, "Physics", "Mr. Ahmed")
classroom3 = Classroom("303", 20, "English", "Miss Ayesha")
classroom4 = Classroom("404", 35, "Chemistry", "Mr. Usman")

# Accessing classroom details
classroom1.class_info()
classroom3.class_info()
````
