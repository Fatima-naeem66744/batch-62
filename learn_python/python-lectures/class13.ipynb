### important points to know

```self``` keyword is the connection between the object and the method.
def is the method
```__init__``` is the constructor function



````python
class teacher():
    def __init__(self , teacher_name : str , teacher_id : int) -> None:
      self.name : str = teacher_name  # atrributes
      self.id : int = teacher_id
      self.oraganization : str = "PUNJAB GROUP OF COLLEGES"

    def speak(self , words) -> None:  
       print(f"{self.name} says '{words}'")

    def teach(self , subject) -> None:    
       print(f"{self.name} is teaching {subject}")

obj1 = teacher("Sir Masood" , 1)
obj2 = teacher("Sir Qasim" , 2) # We can say that the object is the instance of class teacher()

print(obj1.name)
print(obj1.oraganization)
print(obj1.id)
print(obj1.teach("Computer"))
print(obj1.speak("SDLC is the process every software engineer must know"))


print(obj2.name)
print(obj2.oraganization)
print(obj2.id)
print(obj2.teach("GENERATIVE AI"))




````

#### Point to ponder 
Python gives us the speciality of using ```this``` keyword instead of ```self``` but mostly we use ```self``` bcz it is mostly used by community.
