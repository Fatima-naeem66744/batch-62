```python
# multiple indexing
import pandas as pd
value = [1,2,3,4,5]
index : list[list[str]]=  [["a1" , "a1" , "a1" , "b1" , "b1"],
                           ["a" , "b" , "c" , "d" , "e"]]

df = pd.Series(value, index=index , name=  "Student Database" , dtype= "int32")

print(df)
```

## Dataframe

```python
import pandas as pd
s1 = pd.Series([1,2,3,4] , name="student id")

s2 = pd.Series([90,80,70,60] , name="marks")
s3 = pd.Series(["fatima" , "tehreem" , "ali" , "maha"] , name="student name")

df = pd.DataFrame({"student id" : s1  , "marks" : s2 , "student name" :s3 })

print(df)



```

```python
import pandas as pd
s1 = pd.Series([1,2,3,4] , name="student id")

s2 = pd.Series([90,80,70,60] , name="marks")
s3 = pd.Series(["fatima" , "tehreem" , "ali" , "maha"] , name="student name")


df = pd.concat([s1 , s2 , s3] , axis=1)

print(df)



```
