```python
flag : bool = True
num : int = 0
while flag:
    
    print(f"current number is {num}")
    num += 1 
    if num == 10:
        break
```

```python
li : list[int] = [100,200,300]
index : int = 0
while index < len(li):
    print(f"current index is {index} and the index of the list is {li[index]}")
    index += 1
    
```

```python
data = []
flag = True

while flag:
    print("Type 'quit' or 'exit' to end the program")
    name = input("Enter your name: \t")
    education = input("Enter your education: \t")

    if name.lower() in ["exit", "quit"] or education.lower() in ["exit", "quit"]:
        flag = False
        break

    data.append({
        "name": name,
        "education": education
    })

    print(data)  # Display the list after each entry
```
