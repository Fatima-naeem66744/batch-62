```python
from typing import Union

pertype = Union[float, int]
grades = []
percentage: list[pertype] = [88, 99.9, 70]

for per in percentage:
    grade = ""
    if 0 <= per < 30:
        grade = "F"
    elif 30 <= per < 50:
        grade = "C"
    elif 50 <= per < 70:
        grade = "B"
    elif 70 <= per < 90:
        grade = "A"
    elif per >= 90:
        grade = "A+"
    grades.append(grade)

print(percentage)
print(grades)
```

```python
list(zip(percentage, grades , list(range(len(percentage)))))
```
